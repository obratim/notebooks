```
using System.Threading;
using System.Globalization;
```

```
static class ThreadLocalRandom
{
    private static int _seed = (int)(DateTime.Now.Ticks / 1000000000);
    private static readonly ThreadLocal<Random> _random = new ThreadLocal<Random>(() => new Random(Interlocked.Increment(ref _seed)));
    public static Random Value => _random.Value;
}
```

```
var data = Enumerable.Range(0, 10).AsParallel()
    .Select(_ => Enumerable.Range(0, 1000).Select(_ => ThreadLocalRandom.Value.NextDouble()).ToArray())
    .ToArray();
if (data.Select(x => string.Join(" ", x.Select(y => y.ToString(CultureInfo.InvariantCulture)))).GroupBy(x => x).Any(x => x.Count() > 1))
    throw new Exception("Found equal arrays");
else
    "All arrays are uniq".Display();

data
//data.Select(x => string.Join(" ", x.Select(y => y.ToString(CultureInfo.InvariantCulture))))
```
