```polyglot-notebook
var random = new Random();
```

```polyglot-notebook
double RandomFloat(int iterations)
{
    var total = 0.0;
    for (var i = 0; i < iterations; ++i)
    {
        total += random.NextDouble() - random.NextDouble();
    }
    return total / iterations;
}

IEnumerable<double> RandomFloats(int count, int iterations)
{
    for (var i = 0; i < count; ++i)
        yield return RandomFloat(iterations);
}
```

```polyglot-notebook
var data = Enumerable.Range(1, 300).Select(i => i * 5).Select(i => (iterations: i, values: RandomFloats(300, i)));
```
